```python
# Importar librerias 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('goat.jpg')
print("El tama√±o de la imagen es: ", bgr.shape) 

# Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(rgb)
```

```python
# mostrar la imagen es escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.axis('off') # Ocultar los ejes
plt .title('Imagen en escala de grises')
plt.show()
```

```python
#Mostrar los componentes hsv de la imagen 
hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(h, cmap='gray')
plt.title('Componente H')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(s, cmap='gray')
plt.title('Componente S')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(v, cmap='gray')
plt.title('Componente V')
plt.axis('off')
plt.show()
```

```python
#Mostrar ajuste de brillo y contraste
alpha = 1.5  # Contraste
beta = 50    # Brillo
adjusted = cv2.convertScaleAbs(bgr, alpha=alpha, beta=beta)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title('Ajuste de Brillo y Contraste')
plt.axis('off')
plt.show()
```

```python
#Aplicar filtro promedio
kernel_size = (5, 5)
blurred = cv2.blur(bgr, kernel_size)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('Filtro Promedio')
plt.axis('off')
plt.show()

#Aplicar filtro gaussiano
gaussian_blurred = cv2.GaussianBlur(bgr, (5, 5), 0)
plt.imshow(cv2.cvtColor(gaussian_blurred, cv2.COLOR_BGR2RGB))
plt.title('Filtro Gaussiano')
plt.axis('off')
plt.show()

#Aplicar filtro mediana
median_blurred = cv2.medianBlur(bgr, 5)
plt.imshow(cv2.cvtColor(median_blurred, cv2.COLOR_BGR2RGB))
plt.title('Filtro Mediana')
plt.axis('off')
plt.show()

#Aplicar filtro bilateral
bilateral_filtered = cv2.bilateralFilter(bgr, 9, 75, 75)
plt.imshow(cv2.cvtColor(bilateral_filtered, cv2.COLOR_BGR2RGB))
plt.title('Filtro Bilateral')
plt.axis('off')
plt.show()
```

```python
#Redimensionar la imagen mas peque√±a
smaller = cv2.resize(bgr, (0,0), fx=0.5, fy=0.5)
smaller_rgb = cv2.cvtColor(smaller, cv2.COLOR_BGR2RGB)
plt.imshow(smaller_rgb)
plt.axis('off')
plt.title('Imagen Redimensionada Peque√±a')
plt.show()
#Redimensionar la imagen mas grande
larger = cv2.resize(bgr, (0,0), fx=1.5, fy=1.5)
larger_rgb = cv2.cvtColor(larger, cv2.COLOR_BGR2RGB)
plt.imshow(larger_rgb)
plt.axis('off')
plt.title('Imagen Redimensionada Grande')
plt.show()
# Rotar la imagen 45 grados
(h, w) = bgr.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(bgr, M, (w, h))
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_rgb)
plt.axis('off')
plt.title('Imagen Rotada 45 grados')
plt.show()
#transladar la imagen 100 pixeles a la derecha y 50 pixeles hacia abajo
M = np.float32([[1, 0, 100], [0, 1, 50]])
translated = cv2.warpAffine(bgr, M, (w, h))
translated_rgb = cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)
plt.imshow(translated_rgb)
plt.axis('off')
plt.title('Imagen Trasladada')
plt.show()
# Voltear la imagen horizontalmente
flipped = cv2.flip(bgr, 1)
flipped_rgb = cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB)
plt.imshow(flipped_rgb)
plt.axis('off')
plt.title('Imagen Volteada Horizontalmente')
plt.show()
# Voltear la imagen verticalmente
flipped_v = cv2.flip(bgr, 0)
flipped_v_rgb = cv2.cvtColor(flipped_v, cv2.COLOR_BGR2RGB)
plt.imshow(flipped_v_rgb)
plt.axis('off')
plt.title('Imagen Volteada Verticalmente')
plt.show()
```

```python
#aplicar deteccion de bordes con el algoritmo de Canny
edges = cv2.Canny(bgr, 100, 200)
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Detecci√≥n de Bordes con Canny')
plt.show()
#aplicar deteccion de bordes con el algoritmo de Sobel
sobelx = cv2.Sobel(bgr, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(bgr, cv2.CV_64F, 0, 1, ksize=5)
sobel_combined = cv2.magnitude(sobelx, sobely)
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')
plt.title('Detecci√≥n de Bordes con Sobel')
plt.show()
#aplicar deteccion de bordes con el algoritmo de Laplacian
laplacian = cv2.Laplacian(bgr, cv2.CV_64F)
plt.imshow(laplacian, cmap='gray')
plt.axis('off')
plt.title('Detecci√≥n de Bordes con Laplacian')
plt.show()
```

```python
#Probar metodos de deteccion de contornos Harris
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
dst = cv2.dilate(dst, None)
bgr[dst > 0.01 * dst.max()] = [0, 0, 255]
harris_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(harris_rgb)
plt.axis('off')
plt.title('Detecci√≥n de Esquinas con Harris')
plt.show()
#Probar metodos de deteccion de contornos Shi-Tomasi
corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 10)
corners = np.int0(corners)
for i in corners:
    x, y = i.ravel()
    cv2.circle(bgr, (x, y), 3, (0, 255, 0), -1)
shi_tomasi_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(shi_tomasi_rgb)
plt.axis('off')
plt.title('Detecci√≥n de Esquinas con Shi-Tomasi')
plt.show()
```

```python
# detector de formas circulos houghcircles
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1, minDist=20,
                           param1=50, param2=30, minRadius=0, maxRadius=0)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv2.circle(bgr, (i[0], i[1]), i[2], (0, 255, 0), 2)
        cv2.circle(bgr, (i[0], i[1]), 2, (0, 0, 255), 3)
houghcircles_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(houghcircles_rgb)
plt.axis('off')
plt.title('Detecci√≥n de C√≠rculos con HoughCircles')
plt.show()

# contornos principales del objeto de la imagen
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, 0)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(bgr, contours, -1, (0, 255, 0), 3)
contours_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(contours_rgb)
plt.axis('off')
plt.title('Contornos Principales del Objeto')
plt.show()
```

```python
#Segmentar objetos de la imagen usando umbralizacion global y adaptive
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
ret, global_thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
adaptive_thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY, 11, 2)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(global_thresh, cmap='gray')
plt.title('Umbralizaci√≥n Global')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Umbralizaci√≥n Adaptativa')
plt.axis('off')
plt.show()
#Segmentacion por color en espacio HSV
hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
lower_blue = np.array([100, 150, 0])
upper_blue = np.array([140, 255, 255])
mask = cv2.inRange(hsv, lower_blue, upper_blue)
result = cv2.bitwise_and(bgr, bgr, mask=mask)
result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.imshow(result_rgb)
plt.axis('off')
plt.title('Segmentaci√≥n por Color en HSV')
plt.show()
#Segmentacion k-means
Z = bgr.reshape((-1, 3))
Z = np.float32(Z)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 4
ret, label, center = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
segmented_image = res.reshape((bgr.shape))
segmented_image_rgb = cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB)
plt.imshow(segmented_image_rgb)
plt.axis('off')
plt.title('Segmentaci√≥n con K-means')
plt.show()
```
