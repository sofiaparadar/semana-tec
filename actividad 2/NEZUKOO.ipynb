```python
#importar librerias
import cv2
import numpy as np 
import matplotlib.pyplot as pyplot
```

```python
bgr = cv2.imread('nezuko.jpg')
print("El tama√±o de la imagen es: ", bgr.shape)

# Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
pyplot.imshow(rgb)
```

```python
# Mostrar la imagen en escalas grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
pyplot.imshow(gray, cmap='gray')
```

```python
# Mostrar las componentes de color de la imagen
pyplot.figure(figsize=(10,7))
pyplot.subplot(1,3,1)
pyplot.imshow(bgr[:,:,0], cmap='gray')
pyplot.title('Canal Azul')
pyplot.subplot(1,3,2)
pyplot.imshow(bgr[:,:,1], cmap='gray')
pyplot.title('Canal Verde')
pyplot.subplot(1,3,3)
pyplot.imshow(bgr[:,:,2], cmap='gray')
pyplot.title('Canal Rojo')
pyplot.show()
```

```python
# Mostrar las componentes hls de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
pyplot.figure(figsize=(10,7))
pyplot.subplot(1,3,1)
pyplot.imshow(hls[:,:,0], cmap='gray')
pyplot.title('Canal H')
pyplot.subplot(1,3,2)
pyplot.imshow(hls[:,:,1], cmap='gray')
pyplot.title('Canal L')
pyplot.subplot(1,3,3)
pyplot.imshow(hls[:,:,2], cmap='gray')
pyplot.title('Canal S')
pyplot.show()
```

```python
# Ajuste de brillo y contraste
alpha = 1.5 # Contraste
beta = 50   # Brillo
adjusted = cv2.convertScaleAbs(bgr, alpha=alpha, beta=beta)
pyplot.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
pyplot.title('Brillo y Contraste Ajustados')
pyplot.show()
```

```python
# Aplicar filtro Gaussiano
blurred = cv2.GaussianBlur(bgr, (15, 15), 0)
pyplot.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
pyplot.title('Imagen con Filtro Gaussiano')
pyplot.show()
```
